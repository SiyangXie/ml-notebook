# 将字符型转换为数值型

```python
df=pd.DataFrame({'col_one':['1.1','2.2','3.3'],
                 'col_two':['4.4','5.5','6.6'],
                 'col_three':['7.7','8.8','-']})
df
```

```python
df.dtypes
```

```python
df.astype({'col_one':'float','col_two':'float'}).dtypes
```

```python
pd.to_numeric(df.col_three, errors='coerce') # 第三列任何无效数据转换为NaN
```

```python
pd.to_numeric(df.col_three, errors='coerce').fillna(0) # fillna()函数将他们替换成0
```

```python
df = df.apply(pd.to_numeric, errors='coerce').fillna(0) # 一次性对整个df使用fillna
df
```

```python
df.dtypes
```

```python
cols = ['CRIM','ZN']
small_boston=pd.read_csv('boston.csv',usecols=cols)
small_boston.info(memory_usage='deep')
```

```python
# 把continet类型改为category类型，可以再压缩空间，只是boston数据库没这个类型
dtypes={'continent':'category'} 
smaller_boston=pd.read_csv('boston.csv', usecols=cols, dtype=dtypes)
smaller_boston.info(memory_usage='deep')
```

# 按行从多个文件中构建DataFrame

```python
# 假设有3个不同的小数据集stocks1, stocks2, stocks3
stocks1=pd.DataFrame({'Date':['2014-10-03','2014-10-03','2014-10-03'], 'Close':[31.50,112.52,57.42], 'Volume':[14070500,21701800,19189500], 'Symbol':['CSCO','AAPL','MSFT']})
stocks1
```

```python
stocks2=pd.DataFrame({'Date':['2014-10-04','2014-10-04','2014-10-04'], 'Close':[113.00,57.24,31.35], 'Volume':[29736800,20085900,18460400], 'Symbol':['AAPL','MSFT','CSCO']})
stocks2
```

```python
stocks3=pd.DataFrame({'Date':['2014-10-05','2014-10-05','2014-10-05'], 'Close':[57.64,31.59,113.05], 'Volume':[16726400,11808600,21453100], 'Symbol':['MSFT','CSCO','AAPL']})
stocks3
```

```python
# 存入本地
stocks1.to_csv('stocks1.csv',index=False)
stocks2.to_csv('stocks2.csv',index=False)
stocks3.to_csv('stocks3.csv',index=False)
```

```python
# 用golb可以传递某种模式，包括未知字符，这样它会返回符合该事件的文件列表
from glob import glob
stock_files=sorted(glob('stocks*.csv'))
stock_files
```

```python
# 索引值存在重复
pd.concat((pd.read_csv(file) for file in stock_files))
```

```python
# 不存在重复
pd.concat((pd.read_csv(file) for file in stock_files), ignore_index=True)
```
